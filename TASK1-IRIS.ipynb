Data Preprocessing

```python
from sklearn import preprocessing
```

```python
X=iris_data.iloc[:,0:4]
X.head()
```

```python
Y=iris_data['Species']
Y=Y.values
Y[0:5]
```

```python
std=preprocessing.StandardScaler()
X=std.fit_transform(X)
X[0:5]
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.5, random_state = 2)
```

```python
X_train.shape
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
from sklearn.metrics import classification_report
```

```python
lr = LogisticRegression()
knn = KNeighborsClassifier()
dt = DecisionTreeClassifier()
```

```python
models = [lr, knn, dt]
scores = []

for model in models:
  model.fit(X_train, Y_train) # LogisticRegression.fit(X_train, y_train) # Fitting Support Vector Classifer to the Training set
  Y_pred = model.predict(X_test) #LogisticRegression.predict(X_test) # Predicting the Test set results
  scores.append(accuracy_score(Y_test, Y_pred)) # Accuracy on the Test set results  
  print("\n\nAccuracy of " + type(model).__name__ + " is", accuracy_score(Y_test, Y_pred))
  print("\nclassification Report:\n",classification_report(Y_test, Y_pred))
```

```python
results = pd.DataFrame({'Models': ['Logistic Regression', 'K-Nearest Neighbors', 'Decision Tree'],'Accuracy': scores})

results = results.sort_values(by='Accuracy', ascending=False)
print(results)
```
